```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read Dataset
dataset = pd.read_csv('Mall-Customers.csv')
```

```python
dataset
```

```python
# Selects all rows (:) and columns  3 and 4
x = dataset.iloc[:, [3, 4]].values
```

```python
from sklearn.cluster import MeanShift, estimate_bandwidth
from sklearn.datasets import make_blobs
bandwidth = estimate_bandwidth(x, quantile=0.2, n_samples=15)
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
label = ms.fit_predict(x)
supervised=pd.DataFrame(dataset)
supervised['Cluster_group']=label
# Evaluation Metrics(quality of a clustering result) 
from sklearn import metrics
metrics.silhouette_score(x, label, metric='euclidean')
```

```python
supervised
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, legend_out=True)
```
