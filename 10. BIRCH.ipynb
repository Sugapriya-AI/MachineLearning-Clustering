```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall-Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, [3, 4]].values
```

```python
# Perform Elbow Method
from sklearn.cluster import KMeans
list1 = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    list1.append(kmeans.inertia_)
plt.plot(range(1, 11), list1)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import Birch
from sklearn.datasets import make_blobs
brc = Birch(branching_factor=50, n_clusters=5, threshold=0.5, compute_labels=True)
cluster=brc.fit_predict(X)
supervised=pd.DataFrame(dataset)
supervised['Cluster_group']=cluster
# Evaluation Metrics(quality of a clustering result) 
from sklearn import metrics
metrics.silhouette_score(X,cluster, metric='euclidean')
```
